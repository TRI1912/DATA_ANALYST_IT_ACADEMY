N1_EX1: Exploració bàsica amb una API de laboratori.
. Consulta l’API pública JSONPlaceholder utilitzant el mètode GET per obtenir:
Llista de publicacions (/posts)
Llista d’usuaris (/users)
Llista de tasques (/todos)

Farem tres peticions GET:
/posts → publicacions
/users → usuaris
/todos → tasques

```python
#Obtenir la llista de publicacions des de l'API
url_posts = "https://jsonplaceholder.typicode.com/posts"
response_posts = requests.get(url_posts)
```

```python
#Obtenir la llista d'usuris des de l'API
url_users = "https://jsonplaceholder.typicode.com/users"
response_users = requests.get(url_users)
```

```python
#Obtenir la llista de tasques
url_todos = "https://jsonplaceholder.typicode.com/todos"
response_todos = requests.get(url_todos)
```

N1_EX2: 
Mostra per pantalla:
La quantitat total de cada recurs.
El codi d'estat de cada petició. (codi d'estat HTTP)


```python
#Convertir la resposta en format JSON

posts = response_posts.json()
users = response_users.json()
todos = response_todos.json()
```

```python
#Mostrar resultats 
print("POSTS")
print("Quantitat:", len(posts))
print("Codi d'estat:", response_posts.status_code)
print()

print("USERS")
print("Quantitat:", len(users))
print("Codi d'estat:", response_users.status_code)
print()

print("TODOS")
print("Quantitat:", len(todos))
print("Codi d'estat:", response_todos.status_code)
```

N1_EX4: Fes una petició POST per crear una nova publicació fictícia. Inclou el títol, el cos del missatge i un userId.
Mostra la resposta JSON.
Mostra el codi d'estat.

```python
#Petició POST (crear una publicació), creació de la publicació fictícia
nova_publicacio = {
    "title": "La meva primera publicació",
    "body": "Aquest és el contingut de prova",
    "userId": 1
}

```

```python
# Enviar la petició POST

response_post = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=nova_publicacio
)
```

```python
#Mostrar les dades obtingudes i el codi d'estat
print("Resposta JSON:")
print(response_post.json())
print()

print("Codi d'estat:")
print(response_post.status_code)
```

N1_EX5:
Fes una petició PATCH per modificar parcialment una publicació existent.
Mostra la resposta JSON.
Mostra el codi d'estat.

```python
#Petició PATCH (modificació parcial),modificació del títol de la publicació amb ID 1, dades a modificar
dades_patch = {
    "title": "Títol modificat amb PATCH"
}
```

```python
#Enviar la pertició PATCH
response_patch = requests.patch(
    "https://jsonplaceholder.typicode.com/posts/1",
    json=dades_patch
)
```

```python
#mostrar la resposta del servidor
print("Resposta JSON:")
print(response_patch.json())
print()

print("Codi d'estat:")
print(response_patch.status_code)
```

N1_EX6:
Fes una petició DELETE sobre una publicació.
Mostra la resposta JSON.
Mostra el codi d'estat.

```python
# Importar bibliotecas necesarias
import requests
import pandas as pd
```

```python
url_package_search = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search"
params = {
    "q": "Barcelona"     # Cerca lliure
}
response_search = requests.get(url_package_search, params=params)
print("Codi d’estat:", response_search.status_code)

data_search = response_search.json()
```

```python
results = data_search["result"]["results"]

# Mostra els noms + IDs dels primers 5 datasets
for ds in results[:5]:
    print("Nom:", ds["title"])
    print("ID:", ds["id"])
    print("---")
```

```python
dataset_id = results[0]["id"]  # o un altre id que triïs
url_package_show = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_show"
response_show = requests.get(url_package_show, params={"id": dataset_id})

data_show = response_show.json()
```

```python
resources = data_show["result"]["resources"]
for r in resources:
    print("Nom recurs:", r["name"])
    print("Format:", r["format"])
    print("Resource ID:", r["id"])
    print()
```

```python
resource_id = resources[0]["id"]  # substitueix amb el recurs que triïs
url_datastore = "https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search"

params_datastore = {
    "resource_id": resource_id,
    "limit": 200,     # volem almenys 100 registres
}

response_datastore = requests.get(url_datastore, params=params_datastore)
print("Codi d’estat:", response_datastore.status_code)

data_records = response_datastore.json()
```

```python
records = data_records["result"]["records"]
df = pd.DataFrame(records)
df.head()
```

```python
df.to_csv("dades_bcn_export.csv", index=False)
print("Fitxer CSV desat correctament!")
```
