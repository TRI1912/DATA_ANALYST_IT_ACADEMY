```python
def validar_entrada(valor):
    """ Comprova que l'entrada sigui numèrica i retorna el valor flotant. """
    try:
        return float(valor)
    except ValueError:
        print("Error: Si us plau, introdueix un número vàlid.")
        return None
    
def convertir_temp(valor, unitat):
    """ Rep una temperatura i la seva unitat, i retorna un diccionari amb
    amb conversions a altres unitats. """
    unitat = unitat.lower()
    resultats = {}
    if unitat == 'c':
        resultats["Celsius"] = valor
        resultats["Fahrenheit"] = (valor * 9/5) + 32
        resultats["Kelvin"] = valor + 273.15
    elif unitat == 'f':
        celsius = (valor - 32) * 5/9
        resultats["Celsius"] = celsius
        resultats["Fahrenheit"] = valor
        resultats["Kelvin"] = celsius + 273.15
    elif unitat == 'k':
        celsius = valor - 273.15
        resultats["Celsius"] = celsius
        resultats["Fahrenheit"] = (celsius * 9/5) + 32
        resultats["Kelvin"] = valor
    else:
        print("Error: Unitat desconeguda. Utilitza 'C', 'F' o 'K'.")
        return None
    return resultats

def mostrar_resultats(diccionari):
    """Imprimeix de manera ordenada les conversions."""
    print("\nResultat de les conversions:")
    for unitat, valor in diccionari.items():
        print(f"{unitat}: {valor:.2f}")

# Programa principal
valor_usuari = input("Introdueix la temperatura: ")
valor = validar_entrada(valor_usuari)

if valor is not None:
    unitat = input("Introdueix la unitat (C/F/K): ")
    resultats = convertir_temp(valor, unitat)
    if resultats:
        mostrar_resultats(resultats)
```

3. Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
import re
from collections import Counter, defaultdict

def analitzar_text(text):
    # Convertim el text a minúscules i extraiem paraules
    text = text.lower()
    paraules = re.findall(r"[a-zàèéíïòóúüç]+", text)

    # Comptem freqüències de paraules
    freq_paraules = Counter(paraules)

    # Frequències de longituds
    freq_longituds = defaultdict(int)

    for paraula in paraules:
        freq_longituds[len(paraula)] += 1

    # Calculem longitud mitjana
    if len(paraules) > 0:
        llargada_total = sum(len(paraula) for paraula in paraules)
        llargada_mitja = llargada_total / len(paraules)
    else:
        llargada_mitja = 0

    return freq_paraules, freq_longituds, llargada_mitja


# Exemple d'ús
text_usuari = input("Introdueix un text per analitzar: ")
freq_paraules, freq_longituds, llargada_mitja = analitzar_text(text_usuari)

# Mostrar resultats
print("\nFrequència de paraules:")
for paraula, freq in freq_paraules.items():
    print(f"{paraula}: {freq}")

print("\nFrequència de longituds:")
for longitud, freq in freq_longituds.items():
    print(f"{longitud} caràcters: {freq} paraules")

print(f"\nLlargada mitjana de les paraules: {llargada_mitja:.2f} caràcters")
```

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, 
intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria 
d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.


```python
def invertir_diccionari(diccionari):
    #Inverteix un diccionari, intercanviant claus i valors. Si hi ha valors repetits, mostra un missatge d'advertiment
    if not isinstance(diccionari, dict):
        print("Error: has de passar un diccionari vàlid.")
        return None

    valors_repetits = {}
    invertit = {}

    for clau, valor in diccionari.items():
        if valor in invertit:
            # Registra quines claus comparteixen el mateix valor
            valors_repetits.setdefault(valor, [invertit[valor]]).append(clau)
        else:
            invertit[valor] = clau

    # Si s'han detectat valors repetits, avisem l'usuari
    if valors_repetits:
        print("Advertiment: s’han trobat valors repetits!")
        for valor, claus in valors_repetits.items():
            print(f"Valor repetit: {valor} associat a les claus: {claus}")
        return None
    else:
        print("Diccionari invertit correctament.\n")
        return invertit


# Exemple d’ús:
dic1 = {"nom": "Anna", "cognom": "Serra", "edat": "25"}
dic2 = {"a": 1, "b": 2, "c": 1}  # Exemple amb valors repetits

print("Exemple 1 (sense repeticions):")
print(invertir_diccionari(dic1))

print("\nExemple 2 (amb repeticions):")
print(invertir_diccionari(dic2))
```

 Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt".

```python
import os
import re
from collections import defaultdict, Counter
#Llegim el text d'un fitxer i analitzar-lo
ruta = os.path.join(os.getcwd(), 'tu-me-quieres-blanca.txt')
with open(r'C:\Users\tpied\OneDrive\Especialización IT\Sprint_9\tu me quieres blanca.txt', 'r', encoding='utf-8') as fitxer:

    text = fitxer.read().lower()

# Extracció de paraules (incloent lletres catalanes)
paraules = re.findall(r"[a-zàèéíïòóúüç]+", text)

#Calcularem la freqüència de paraules
freq = Counter(paraules)

#Ordenem per lletres inicials amb un diccionari
diccionari = defaultdict(dict)
for paraula in freq:
    inicial = paraula[0].upper()
    if 'A' <= inicial <= 'Z':
        diccionari[inicial][paraula] = freq[paraula]

# Ordenem inicials i paraules
resultat = {lletra: {p: diccionari[lletra][p] for p in sorted(diccionari[lletra])}
            for lletra in sorted(diccionari)}

# Mostrem el resultado
for lletra in resultat:
    print(f"{lletra}:")
    for paraula, compte in resultat[lletra].items():
        print(f"  {paraula}: {compte}")
```

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir: llista ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']

```python
def conversion(dades):
    convertibles = []
    no_convertibles = []

    for item in dades:
        # L'exercici és una colecció. Però també funcionen tuples i altres llistes com a entrada, iterar per dins.
        if isinstance(item, (list, tuple)):
            for x in item:
                try:
                    convertibles.append(float(x))
                except:
                    no_convertibles.append(x)
        else:
            try:
                convertibles.append(float(item))
            except:
                no_convertibles.append(item)
    return (convertibles, no_convertibles)


resultat = conversion(['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']])
print(resultat)
```

Generador de contrasenyes

```python
import numpy as np

def crear_contrasenya(longitud: int,
                       majuscules: bool = True,
                       minuscules: bool = True,
                       numeros: bool = True,
                       signes: bool = False) -> str:
    """Genera una contrasenya aleatòria amb els paràmetres especificats."""

    # Conjunts de caràcters
    C_MAJ = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    C_MIN = "abcdefghijklmnopqrstuvwxyz"
    C_NUM = "0123456789"
    C_SIG = ",-$?@!%&/()[]{}*#"

    # Construcció del conjunt total de caràcters
    conjunts = []
    if majuscules: conjunts.append(C_MAJ)
    if minuscules: conjunts.append(C_MIN)
    if numeros:    conjunts.append(C_NUM)
    if signes:     conjunts.append(C_SIG)

    if not conjunts:
        raise ValueError("Has d'activar com a mínim un tipus de caràcter!")

    # Assegurem que la longitud és suficient
    if longitud < len(conjunts):
        raise ValueError("La longitud és massa curta per incloure tots els tipus seleccionats.")

    # Garantim almenys un caràcter de cada tipus
    contrasenya_chars = [np.random.choice(list(c)) for c in conjunts]

    # Caràcters restants
    tots_caracters = "".join(conjunts)
    restants = np.random.choice(list(tots_caracters), size=longitud - len(conjunts))

    # Combinar i barrejar
    contrasenya_chars.extend(restants)
    np.random.shuffle(contrasenya_chars)

    return "".join(contrasenya_chars)

# ÚS
pwd = crear_contrasenya(10, True, True, True, True)
print(pwd)

```

Processament de dades simple
Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

```python
import re
from collections import defaultdict

# Per llegir el fitxer 
fitxer = "C:\\Users\\tpied\\OneDrive\\Especialización IT\\Sprint_9\\historic partits.txt"

with open(fitxer, "r", encoding="utf-8") as f:
    linies = f.read().strip().splitlines()

# Estructurem les dades
gols_favor = defaultdict(int)
gols_contra = defaultdict(int)
punts = defaultdict(int)

# Revisem cada línia de fitxer que es correspon a un partit.
for linia in linies:
    # Format Exemple:  "Manlleu    0-1    Granollers"
    m = re.match(r"(.+?)\s+(\d+)-(\d+)\s+(.+)", linia)
    if not m:
        continue
    equip_casa, gf, gc, equip_fora = m.groups()
    gf, gc = int(gf), int(gc)

    # Actualitzem els gols segons el partit:
    gols_favor[equip_casa] += gf
    gols_contra[equip_casa] += gc
    gols_favor[equip_fora] += gc
    gols_contra[equip_fora] += gf

    # Assignació de punts segons el resultat:
    if gf > gc:
        punts[equip_casa] += 3
    elif gc > gf:
        punts[equip_fora] += 3
    else:
        punts[equip_casa] += 1
        punts[equip_fora] += 1

# Resultats

print("\n--- GOLS MARCATS PER EQUIP ---")
for equip, gols in sorted(gols_favor.items(), key=lambda x: -x[1]):
    print(f"{equip}: {gols} gols")

equip_mes_golejador = max(gols_favor, key=gols_favor.get)
equip_mes_golejat = max(gols_contra, key=gols_contra.get)

print("\n--- Equip més golejador ---")
print(f"{equip_mes_golejador} ({gols_favor[equip_mes_golejador]} gols)")

print("\n--- Equip més golejat ---")
print(f"{equip_mes_golejat} ({gols_contra[equip_mes_golejat]} gols encaixats)")

print("\n--- CLASSIFICACIÓ (ordenada per punts) ---")
classificacio = sorted(punts.items(), key=lambda x: -x[1])
for pos, (equip, p) in enumerate(classificacio, start=1):
    print(f"{pos}. {equip}: {p} punts")
```
