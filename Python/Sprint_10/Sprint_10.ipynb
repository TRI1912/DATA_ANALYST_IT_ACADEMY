**NIVELL_I_EX_3**
Junta les columnes Fills i No Fills en una sola columna, utilitzant el mètode .apply() i definint una funció que resolgui el problema. La columna nova ha de dir-se "Fills" i prendre els valors "Sí" o "No".

```python
def combinar_fills(row):
    if pd.notna(row["Fills"]) and bool(row["Fills"]) is True:
        return "Sí"
    if pd.notna(row["No Fills"]) and bool(row["No Fills"]) is True:
        return "No"
    return pd.NA

df3 = df2.copy()
df3["Fills"] = df3.apply(combinar_fills, axis=1)
df3 = df3.drop(columns=["No Fills"])

print(df3[["Nom", "Cognoms", "Fills"]].head())
```

**NIVELL_I_EX_6**
Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.
Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python
from datetime import date

# Primer crearem la columna datetime a partir de Dia / Mes / Any
df3["Data_naixement"] = pd.to_datetime(
    dict(year=df3["Any"], month=df3["Mes"], day=df3["Dia"])
)

def calcular_edat(data_naixement, avui=None):
    if pd.isna(data_naixement):
        return pd.NA
    if avui is None:
        avui = date.today()
    anys = avui.year - data_naixement.year
    if (avui.month, avui.day) < (data_naixement.month, data_naixement.day):
        anys -= 1
    return anys

df3["Edat"] = df3["Data_naixement"].apply(calcular_edat)

print(df3[["Nom", "Cognoms", "Data_naixement", "Edat"]].head())
```

```python
# Recuperarem DNI per poder fer el merge
df2 = df2.reset_index()

df_increment = pd.DataFrame({
    "Grup Professional": ["Grup A", "Grup B", "Grup C", "Grup D"],
    "Increment": ["5%", "3,5%", "2%", "8%"]
})

# Convertirem text a percentatge numèric
df_increment["Increment"] = (
    df_increment["Increment"]
    .str.replace("%", "", regex=False)
    .str.replace(",", ".", regex=False)
    .astype(float) / 100
)

# Afegirem Increment a df2 pel Grup Professional
df2 = df2.merge(df_increment, on="Grup Professional", how="left")

# Convertirem salari textual a numèric
def parse_salari(x):
    if isinstance(x, str):
        return float(x.replace(".", "").replace(" €", "").replace(",", "."))
    return float(x)

df2["Salari mensual"] = df2["Salari mensual"].apply(parse_salari)

# Actualitzarem salari amb increment
df2["Salari mensual"] = df2["Salari mensual"] * (1 + df2["Increment"])

# Arrodonirem
df2["Salari mensual"] = df2["Salari mensual"].round(2)

# Tornar a posar DNI com a índex
df2 = df2.set_index("DNI")

df2.head()
```

```python
# 1) Exportar un fitxer per a cada Grup Professional
for grup, subdf in df2.groupby("Grup Professional"):
    # Treurem espais del nom del grup per construir el nom del fitxer
    nom_grup = grup.replace(" ", "")
    nom_fitxer = f"dades_{nom_grup}.xlsx"
    subdf.to_excel(nom_fitxer)

#  CSV:
#     nom_fitxer = f"dades_{nom_grup}.csv"
#     subdf.to_csv(nom_fitxer)

# 2) Crearem el DataFrame resum amb recompte, sou mig i edat mediana
# Utilitzem df3 perquè és el DataFrame que conté la columna 'Edat' i el salari numèric 'Salari_num'.
df_resum = (
    df3
    .groupby("Grup Professional")
    .agg(
        Treballadors=("Grup Professional", "count"),
        Sou_mig=("Salari_num", "mean"),
        Edat_mediana=("Edat", "median")
    )
)

# Arrodoniments menors per presentació
df_resum["Sou_mig"] = df_resum["Sou_mig"].round(2)
df_resum["Edat_mediana"] = df_resum["Edat_mediana"].round(1)

# 3) Exportarem el DataFrame resum
df_resum.to_excel("resum_grups_professionals.xlsx")
df_resum.to_csv("resum_grups_professionals.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_all_columns(df):
    for col in df.columns:
        s = df[col].dropna()
        if s.empty:
            continue

        plt.figure()

        # Variables numèriques
        if pd.api.types.is_numeric_dtype(s):
            sns.histplot(s, kde=False)
            plt.title(f"Histograma de {col}")
            plt.xlabel(col)
            plt.ylabel("Freqüència")

        # Variables de dates
        elif pd.api.types.is_datetime64_any_dtype(s):
            anys = s.dt.year.value_counts().sort_index()
            anys.plot(kind="bar")
            plt.title(f"Anys més freqüents ({col})")
            plt.xlabel("Any")
            plt.ylabel("Comptatge")

        # Variables categòriques
        else:
            freq = s.value_counts().head(10)
            freq.plot(kind="bar")
            plt.title(f"Valors més freqüents ({col})")
            plt.xlabel(col)
            plt.ylabel("Comptatge")

        plt.tight_layout()
        plt.show()
```

```python
iris = sns.load_dataset("iris")
plot_all_columns(iris)
iris
```

```python
penguins = sns.load_dataset("penguins")
plot_all_columns(penguins)
penguins
```

```python
import pandas as pd

# Carregarem l’Excel (full amb distàncies)
df_dist = pd.read_excel(r"C:\Users\tpied\OneDrive\Especialización IT\Sprint_10\matriu_distancies.xlsx", sheet_name="Distancia (kilómetros)")

# Posarem els noms de ciutat com a índex
df_dist = df_dist.rename(columns={"Unnamed: 0": "Ciutat"}).set_index("Ciutat")

# Eliminarem ciutats que no es poden fer en cotxe
df_dist = df_dist.drop(index=["Las Palmas de Gran Canaria", "Palma"],
                       columns=["Las Palmas de Gran Canaria", "Palma"])
```

```python
df_dist
```

```python
def ruta_vei_mes_proper(df_dist, origen):
    ciutats = list(df_dist.index)
    if origen not in ciutats:
        raise ValueError("La ciutat d'origen no és a la matriu")

    visitades = [origen]
    actual = origen
    distancia_total = 0.0

    while len(visitades) < len(ciutats):
        # ciutats encara no visitades
        no_visitades = [c for c in ciutats if c not in visitades]
        # distàncies des de la ciutat actual a les no visitades
        distancies = df_dist.loc[actual, no_visitades]
        propera = distancies.idxmin()
        d = distancies.min()

        visitades.append(propera)
        distancia_total += d
        actual = propera

    return visitades, distancia_total
```

```python
ruta, dist_total = ruta_vei_mes_proper(df_dist, origen="Barcelona")
print("Ruta:", ruta)
print("Distància total:", dist_total)
```

```python
def millor_origen(df_dist):
    resultats = {}
    for ciutat in df_dist.index:
        _, dist_total = ruta_vei_mes_proper(df_dist, origen=ciutat)
        resultats[ciutat] = dist_total

    # trobar ciutat amb distància mínima
    millor_ciutat = min(resultats, key=resultats.get)
    return millor_ciutat, resultats[millor_ciutat], resultats

millor_ciutat, millor_dist, totes = millor_origen(df_dist)
print("Millor ciutat d'origen:", millor_ciutat)
print("Distància total amb aquesta ciutat:", millor_dist)
```
